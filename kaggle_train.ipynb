```python
# ── 4. Restore checkpoint from previous session (if any) ─────────────────────
# If you saved a checkpoint at the end of a previous session, upload it to
# Kaggle as a dataset and copy it here before training starts.
import shutil, os

CHECKPOINT_INPUT = "/kaggle/input/omiya-checkpoint/checkpoint_latest.pt"  # adjust dataset name
CHECKPOINT_DST   = "runs/kaggle_run/checkpoint_latest.pt"

if os.path.exists(CHECKPOINT_INPUT):
    os.makedirs("runs/kaggle_run", exist_ok=True)
    shutil.copy(CHECKPOINT_INPUT, CHECKPOINT_DST)
    print(f"Restored checkpoint from {CHECKPOINT_INPUT}")
else:
    print("No prior checkpoint found — starting fresh.")
```

```python
# ── 6. Save outputs before session ends ──────────────────────────────────────
# Run this cell manually before your session times out.
import shutil, os

OUT = "/kaggle/working"
RUN = "runs/kaggle_run"

for fname in ["checkpoint_latest.pt", "training_progress.png", "training_summary.csv",
              "policy_last.pt", "critic_last.pt"]:
    src = os.path.join(RUN, fname)
    if os.path.exists(src):
        shutil.copy(src, os.path.join(OUT, fname))
        print(f"Saved {fname}")

print("\nDownload these from the Output tab on the right.")
```
